# **Data preprocessing**

```python
from sklearn.preprocessing import LabelEncoder
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
df = df.apply(lambda x: object_to_int(x))
df.head()
```

```python
plt.figure(figsize=(14,7))
df.corr()['Churn'].sort_values(ascending = False)
```

```python
df.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
plt.title("Correlation of Features with Churn")
plt.show()
```

```python
X = df.drop(columns = ['Churn'])
y = df['Churn'].values
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0, stratify=y)
```

```python
def distplot(feature, frame, color='r'):
    plt.figure(figsize=(8,3))
    plt.title("Distribution for {}".format(feature))
    ax = sns.distplot(frame[feature], color= color)
```

```python
num_cols = ["tenure", 'MonthlyCharges', 'TotalCharges']
for feat in num_cols: distplot(feat, df)
```

# **Standardizing numeric attributes**

```python
from sklearn.preprocessing import StandardScaler # Import StandardScaler

df_std = pd.DataFrame(StandardScaler().fit_transform(df[num_cols].astype('float64')),
                       columns=num_cols)
for feat in numerical_cols: distplot(feat, df_std, color='c')
```

```python
# Divide the columns into 3 categories, one ofor standardisation, one for label encoding and one for one hot encoding
cat_cols_ohe =['PaymentMethod', 'Contract', 'InternetService'] # those that need one-hot encoding
cat_cols_le = list(set(X_train.columns)- set(num_cols) - set(cat_cols_ohe)) #those that need label encoding
```

```python
scaler= StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

# **Logistic Regression Model**

```python
# Import necessary libraries

#This line imports the LogisticRegression class from the sklearn.linear_model library.
#LogisticRegression is a machine learning model used for binary or multiclass classification, using logistic regression to predict the probabilities of classes.
from sklearn.linear_model import LogisticRegression

#Split the data into training set and test set.
from sklearn.model_selection import train_test_split

#Evaluate model accuracy.
from sklearn.metrics import accuracy_score
```

```python
# Step 1: Initialize and train the Logistic Regression model

#Initialize the Logistic Regression model
lr_model = LogisticRegression()
#Train the model
lr_model.fit(X_train,y_train)
```

```python
# Step 2: Calculate the model's accuracy

#lr_model.score(X_test, y_test) calculates the accuracy of the lr_model on the test data (X_test and y_test)
accuracy_lr = lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

```python
# Step 3: Predict and create assessment report

# Import classification_report
from sklearn.metrics import classification_report

#lr_model.predict(X_test) uses the trained lr_model to predict the label (y) for the data in X_test.
lr_pred= lr_model.predict(X_test)

#classification_report(y_test, lr_pred) generates an evaluation report of the model's performance metrics, comparing the actual label (y_test) with the predicted label (lr_pred).
report = classification_report(y_test,lr_pred)
print(report)
```

```python
# Step 4: Draw the Confusion Matrix

from sklearn.metrics import confusion_matrix # Import confusion_matrix

#plt.figure(figsize=(4,3)) sets the size of the chart (4 units wide and 3 units high).
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, lr_pred),
                annot=True,fmt = "d",linecolor="k",linewidths=3)

#plt.title(...) adds the title "LOGISTIC REGRESSION CONFUSION MATRIX" with a font size of 14 to describe the chart type.
plt.title("LOGISTIC REGRESSION CONFUSION MATRIX",fontsize=14)
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Predict the probability for the positive class (class 1)
y_pred_prob = lr_model.predict_proba(X_test)[:, 1]

# Compute FPR, TPR, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Calculate AUC
roc_auc = auc(fpr, tpr)

# Plot the baseline diagonal line (random guess)
plt.plot([0, 1], [0, 1], 'k--')

# Plot the actual ROC curve for the logistic regression model
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})', color='r')

# Label the axes and set the title for the plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve', fontsize=16)
plt.legend(loc='lower right')  # Display legend with AUC value

# Show the plot
plt.show()
```
